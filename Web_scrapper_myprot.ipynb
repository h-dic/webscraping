```python
import requests
page_1 = requests.get("https://fr.myprotein.com/nutrition/protein.list")
page_2 = requests.get("https://fr.myprotein.com/nutrition/protein.list?pageNumber=2")
```

```python
pages = page_1.content + page_2.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(pages,'html.parser')
```

```python
proteines_name = soup.find_all('h2', class_ ='athenaProductBlock_productName')
proteines_price_from = soup.find_all('span', class_='athenaProductBlock_fromValue')
proteines_price_value = soup.find_all('span', class_='athenaProductBlock_priceValue')
```

```python
for item in list(proteines_price_from):
    item['class']='price'

for item in list(proteines_price_value):
    item['class']='price'
    
proteines_price = soup.find_all('span', class_='price') 
```

```python
list_proteines_name = [item.get_text()[1:-1] for item in list(proteines_name)]
list_proteines_price = [item.get_text()[:-2] for item in list(proteines_price)]
```

```python
print(list_proteines_name)
print(list_proteines_price)
```

```python
len(list_proteines_price)
```

```python
import pandas as pd
```

```python
proteins = pd.DataFrame({
    "name":list_proteines_name,
    "price":list_proteines_price
})
```
